#Using LLM in Python

This is a short practice on how we can use Hugging Face Transformers library transformers to generate text using a pre-trained model.

```python
# 1. Install the library
!pip install transformers
```

```python
from transformers import pipeline
```

```python
# 2. Load a small pre-trained text generation model
generator = pipeline("text-generation", model="gpt2")
```

```python
# 3. Generate text
prompt = "My name is"
result = generator(prompt, max_length=30, num_return_sequences=2)
```

```python
print(result[0]['generated_text'])
```
